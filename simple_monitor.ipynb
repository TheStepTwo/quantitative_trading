```python
def stockPriceColor2(row):
    rowColorList = []
    for index, value in enumerate(row):
        if a.columns.get_loc("price") != index:
            rowColorList.append("");
        else: 
            if value > row.lastPrice:
                rowColorList.append("background: red");
            elif value < row.lastPrice:
                rowColorList.append("background: #2EFA84");
            else:
                rowColorList.append("");
    return rowColorList
```

```python
import time
from IPython.display import clear_output
timePeroid = 5.0;
starttime=time.time()
quotation = easyquotation.use("hkquote")
monList = [ '03175' , '03323' , '07200' , '00788' , '00939' , '01928' , '01186' , '09988' , '03998' , '01398' , '01810' , '00241' , '03988' , '00700', '09928' , '01755' , '06186' , '07500' , '07522' , '07331' , '00728' ]
myList = ['07500' , '09988']
while True:
    data = quotation.real(monList)
    a = pd.DataFrame(data)
    a = a.T
    clear_output(wait=True)
    print('----')
    clear_output(wait=True)
    a['upDown'] = np.where(a['price'] - a['lastPrice'] > 0 , '↑', '↓')
    a['%'] = (a['price'] - a['lastPrice'] ) / a['lastPrice']  * 100
    a.sort_values(by='%', ascending=True , inplace = True )
    a['%'] = a['%'].astype(float).round(2).astype(str) + '%'
    displayer = a.style.apply(stockPriceColor2,axis=1)
    display(displayer)
    time.sleep( timePeroid - ((time.time() - starttime) % timePeroid ))
```

```python
import easyquotation
import numpy as np
quotation  = easyquotation.use("daykline")
trade_stock = '03998'
data = quotation.real([trade_stock])
# print(data)
type(data[trade_stock])
data[trade_stock]
# dfKLine = pd.DataFrame(data[trade_stock])
dfKLine = pd.DataFrame(data[trade_stock])
dfKLine.columns =['date', 'openPrice' , 'lastPrice' , 'high' , 'low' , 'amount' , 'remark1' , 'remark2' , 'remark3' ]
# dfKLine.set_index('date' , inplace = True)
dfKLine.sort_values( by = 'date' , inplace=True , ascending=False)
dfKLine.amount = dfKLine.amount.astype('float64')
dfKLine.remark3 = dfKLine.remark3.astype('float64')
dfKLine.lastPrice = dfKLine.lastPrice.astype('float64')
dfKLine.openPrice = dfKLine.openPrice.astype('float64')
# dfKLine['%'] = (dfKLine['lastPrice'] - dfKLine['openPrice'] ) / dfKLine['openPrice']  * 100
dfKLine['Change'] = dfKLine.lastPrice.astype('float64') - dfKLine.lastPrice.shift(-1).astype('float64')
dfKLine['%'] = dfKLine['Change']/ dfKLine.lastPrice.shift(-1).astype('float64') * 100
dfKLine
```

```python
import plotly.express as px
import plotly.graph_objects as go

df = dfKLine

fig = go.Figure(data=[go.Candlestick(x=df['date'],
                open=df['openPrice'],
                high=df['high'],
                low=df['low'],
                close=df['lastPrice'])])
fig.update_layout(
    title= {
        'text': 'hi',
        'y':0.6,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
      font=dict(
        family="Courier New, monospace",
        size=20,
        color="#7f7f7f"
    )
    )
fig.show()
```
